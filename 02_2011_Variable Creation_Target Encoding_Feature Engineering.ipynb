```python
import pandas as pd
import datetime
import numpy as np
data=pd.read_csv('01_variable_creation.csv')
```

```python
data=data.drop('Unnamed: 0',axis=1)
```

```python
data
```

```python
data['Date'] = [datetime.datetime.strptime(date, '%Y-%m-%d') for date in data['Date']]
data['quarter']=data['Date'].dt.quarter
```

```python
new_data=data.copy()
```

```python
keys24= ['temp_max_24','temp_max_24','temp_max_24','temp_max_24','temp_max_24']
keys12= ['temp_max_12','temp_max_12','temp_max_12','temp_max_12','temp_max_12']
keys6= ['temp_max_6','temp_max_6','temp_max_6','temp_max_6','temp_max_6']
max24={}
max12={}
max6={}
for i in range(5):
    col='Temperature'
    rolling_max24 = data[col].rolling(window=24).max()
    max24[keys24[i]]=rolling_max24
    rolling_max12 = data[col].rolling(window=12).max()
    max12[keys12[i]]=rolling_max12
    rolling_max6 = data[col].rolling(window=6).max()
    max6[keys6[i]]=rolling_max6
```

```python
#dict to dataframe
max6_df=pd.DataFrame.from_dict(max6)
max12_df=pd.DataFrame.from_dict(max12)
max24_df=pd.DataFrame.from_dict(max24)
```

```python
new_data=pd.merge(new_data,max6_df,left_on=new_data.index,right_on=max6_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,max12_df,left_on=new_data.index,right_on=max12_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,max24_df,left_on=new_data.index,right_on=max24_df.index,how='inner').drop('key_0',axis=1)
```

```python
keys24= ['temp_min_24','temp_min_24','temp_min_24','temp_min_24','temp_min_24']
keys12= ['temp_min_12','temp_min_12','temp_min_12','temp_min_12','temp_min_12']
keys6= ['temp_min_6','temp_min_6','temp_min_6','temp_min_6','temp_min_6']
min24={}
min12={}
min6={}
for i in range(5):
    col='Temperature'
    rolling_min24 = data[col].rolling(window=24).min()
    min24[keys24[i]]=rolling_min24
    rolling_min12 = data[col].rolling(window=12).min()
    min12[keys12[i]]=rolling_min12
    rolling_min6 = data[col].rolling(window=6).min()
    min6[keys6[i]]=rolling_min6
```

```python
#dict to dataframe
min6_df=pd.DataFrame.from_dict(min6)
min12_df=pd.DataFrame.from_dict(min12)
min24_df=pd.DataFrame.from_dict(min24)
```

```python
new_data=pd.merge(new_data,min6_df,left_on=new_data.index,right_on=min6_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,min12_df,left_on=new_data.index,right_on=min12_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,min24_df,left_on=new_data.index,right_on=min24_df.index,how='inner').drop('key_0',axis=1)
```

```python
keys24= ['temp_std_24','temp_std_24','temp_std_24','temp_std_24','temp_std_24']
keys12= ['temp_std_12','temp_std_12','temp_std_12','temp_std_12','temp_std_12']
keys6= ['temp_std_6','temp_std_6','temp_std_6','temp_std_6','temp_std_6']
std24={}
std12={}
std6={}
for i in range(5):
    col='Temperature'
    rolling_std24 = data[col].rolling(window=24).std()
    std24[keys24[i]]=rolling_std24
    rolling_std12 = data[col].rolling(window=12).std()
    std12[keys12[i]]=rolling_std12
    rolling_std6 = data[col].rolling(window=6).std()
    std6[keys6[i]]=rolling_std6
```

```python
#dict to dataframe
std6_df=pd.DataFrame.from_dict(std6)
std12_df=pd.DataFrame.from_dict(std12)
std24_df=pd.DataFrame.from_dict(std24)
```

```python
new_data=pd.merge(new_data,std6_df,left_on=new_data.index,right_on=std6_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,std12_df,left_on=new_data.index,right_on=std12_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,std24_df,left_on=new_data.index,right_on=std24_df.index,how='inner').drop('key_0',axis=1)
```

```python
keys24= ['temp_mean_24','temp_mean_24','temp_mean_24','temp_mean_24','temp_mean_24']
keys12= ['temp_mean_12','temp_mean_12','temp_mean_12','temp_mean_12','temp_mean_12']
keys6= ['temp_mean_6','temp_mean_6','temp_mean_6','temp_mean_6','temp_mean_6']
mean24={}
mean12={}
mean6={}
for i in range(5):
    col='Temperature'
    rolling_mean24 = data[col].rolling(window=24).mean()
    mean24[keys24[i]]=rolling_mean24
    rolling_mean12 = data[col].rolling(window=12).mean()
    mean12[keys12[i]]=rolling_mean12
    rolling_mean6 = data[col].rolling(window=6).mean()
    mean6[keys6[i]]=rolling_mean6
```

```python
#dict to dataframe
mean6_df=pd.DataFrame.from_dict(mean6)
mean12_df=pd.DataFrame.from_dict(mean12)
mean24_df=pd.DataFrame.from_dict(mean24)
```

```python
new_data=pd.merge(new_data,mean6_df,left_on=new_data.index,right_on=mean6_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,mean12_df,left_on=new_data.index,right_on=mean12_df.index,how='inner').drop('key_0',axis=1)
new_data=pd.merge(new_data,mean24_df,left_on=new_data.index,right_on=mean24_df.index,how='inner').drop('key_0',axis=1)
```

```python
cat_variable=['Hour','month','quarter','weekday','month.week','Time_of_day','Day_Night_Time']
new_data_join=new_data
import category_encoders as ce
for col in cat_variable:
    X = new_data_join[col]
    y = new_data_join['Load']

    te = ce.target_encoder.TargetEncoder(cols=col)
    te.fit(X,y)
    X_TE = te.transform(X,y)
    new_data_join=pd.merge(new_data_join, X_TE, left_on=new_data_join.index, right_on=X_TE.index, how='inner',suffixes=('', '_Mean_Encoding'))
    new_data_join=new_data_join.drop('key_0',axis=1)
```
