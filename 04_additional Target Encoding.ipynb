```python
import pandas as p

data = pd.read_csv('03_2010_v3_variable_creation.csv')
```

```python
data['trend']=data['X']
```

```python
data = data.drop(['Unnamed: 0','X'],axis=1)
```

```python
import numpy as np
new_data=data.copy()
new_data.loc[(new_data['Date']<'2012-01-01') & (new_data['Date']>='2011-01-01'),'Load']=np.nan
new_data[(new_data['Date']<'2012-01-01') & (new_data['Date']>='2011-01-01')]['Load']
```

```python
cat_variable=['Time_of_day','Day_Night_Time']
new_data_join=new_data
import category_encoders as ce
for col in cat_variable:
    X = new_data_join[col]
    y = new_data_join['Load']

    te = ce.target_encoder.TargetEncoder(cols=col)
    te.fit(X,y)
    X_TE = te.transform(X,y)
    new_data_join=pd.merge(new_data_join, X_TE, left_on=new_data_join.index, right_on=X_TE.index, how='inner',suffixes=('', '_Mean_Encoding'))
    new_data_join=new_data_join.drop('key_0',axis=1)
```

```python
new_data_join.loc[(new_data_join['Date']<'2012-01-01') & (new_data_join['Date']>='2011-01-01'),'Load']\
=data.loc[(data['Date']<'2012-01-01') & (data['Date']>='2011-01-01'),'Load']
```
